```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns',None)
sns.set_theme(color_codes=True)
```

```python
file_path = 'C:\\Users\\Admin\\Data Science\\Real Time projects\\Telecom_Customer_Churn\\Preprocessed_data.csv'
df = pd.read_csv(file_path)
df
```

```python
X = df.drop('Customer Status',axis=1)
```

```python
y = df['Customer Status']
```

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.tree import DecisionTreeClassifier
gridtree = DecisionTreeClassifier()
gridtree
```

```python
gridtree.get_params()
```

```python
param_grid = {'criterion' :['gini','entropy'],
              'max_depth' : [3,4,5,6,7,8] ,
              'min_samples_leaf' : [2,3,4],
              'min_samples_split' : [1,2,3,4],
              'random_state' : [0,42]}
```

```python
grid_search = GridSearchCV(gridtree,param_grid,
                           scoring='accuracy',
                           cv = 5,
                           verbose=True)
```

```python
grid_search
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 random_state=0,
                                                 test_size=0.2)
```

```python
import time
start = time.time()
grid_search.fit(X_train,y_train)
end = time.time()
print('The time taken =',(end-start))
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search_proba = grid_search.predict_proba(X_test)[:,1]
from sklearn.metrics import roc_curve
fpr,tpr,threshold = roc_curve(y_test,grid_search_proba)
plt.plot(fpr,tpr)
plt.show()
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
acccuracy = round(accuracy_score(y_test,y_pred)*100,2)
precision = round(precision_score(y_test,y_pred)*100,2)
recall = round(recall_score(y_test,y_pred)*100,2)
f1_sco = round(f1_score(y_test,y_pred)*100,2)
print('acccuracy is ',acccuracy)
print('precision is ',precision)
print('recall is ',recall)
print('f1_score',f1_sco)

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cmt = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cmt,
                              display_labels=[False,True])
disp.plot()
plt.grid()
plt.show()
```
